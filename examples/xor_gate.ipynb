```python
''' importing the MLP from bishograd-neural-network '''
from bishograd.nn import MLP
```

```python
'''
sample neural-network to solve XOR Logic Gate
The XOR Gate works with 2 inputs: x1 & x2 and produces an output y
when x1 & x2 are opposite (0 & 1 or vice-versa) the outcome 'y' is 1 else 0
'''
## all four possible input combinations
xor_inputs = [
    [1.0,0.0],
    [1.0,1.0,],
    [0.0,1.0],
    [0.0,0.0,]
]

## outcome for given combinations
y_outputs = [1.0,0.0,1.0,0.0]
```

```python
'''
Making a simple neural network with MLP, having input size of 2, 2 hidden layer
with 4 neuron each and 1 output layer with 1 neuron along with activation function for each layer.
'''
nn = MLP(2,[4,4,1],['relu','relu','sigmoid'])
```

```python
'''
Training the model with following hyperparameters:
epochs = 1000
learning rate = 0.2
'''
nn.train(1000,0.2,xor_inputs,y_outputs,stats=True)
```

```python
## testing our model 
x_test =[
     [0,1],
     [1,0],
     [0,0],
     [1,1]
]
for i in x_test:
     y_test = nn(x=i)
     print(f'Input: {i} --- Prediction: {y_test.data:.2f} --- Correct Output: {i[0]^i[1]}')
```
